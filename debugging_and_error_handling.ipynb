Given an array `ar = ['AI',1,'is',2,'the',2.4,'new',True,'electricity', {"MasterCourse": "bridge between academia and industry"}]`, count the number of different data types.

```python
ar = ['AI',1,'is',2,'the',2.4,'new',True,'electricity', {"MasterCourse": "bridge between academia and industry"}]

count_types = {}
for element in ar:
  elem_type = type(element)
  if elem_type in count_types.keys():
    count_types[elem_type] += 1
  else:
    count_types[elem_type] = 1
count_types
```

Given an array `ar = ['AI',1,'is',2,'the',2.4,'new',True,'electricity', {"MasterCourse": "bridge between academia and industry"}]`, find the sum of the numbers

```python
ar = ['AI',1,'is',2,'the',2.4,'new',True,'electricity', {"MasterCourse": "bridge between academia and industry"}]

sum = 0
for element in ar:
  try:
    sum += element
  except TypeError:
    continue

sum
```
